```python
import os
import pandas as pd
```

```python
events_adapt_dir = '/Users/carinakauf/repos/lm-event-knowledge/human_ratings/2_EventsAdapt_plausibility/results_raw'
events_rev_dir = '/Users/carinakauf/repos/lm-event-knowledge/human_ratings/1_EventsRev_plausibility/results_raw'
```

```python
def get_comments(study='EventsAdapt'):
    if study == 'EventsAdapt':
        data_dir = events_adapt_dir
    elif study == 'EventsRev':
        data_dir = events_rev_dir
    frames = []
    for filename in os.listdir(data_dir):
        df = pd.read_csv(os.path.join(data_dir,filename))
        frames.append(df)
    full_df = pd.concat(frames)
    
    participant_comments = list(full_df['Answer.answer'])
    participant_comments = [elm for elm in participant_comments if elm != '{}']
    
    for elm in participant_comments:
        print(elm)
        
    with open(f'{study}_participant_comments.txt', 'w') as f:
        for item in participant_comments:
            f.write("%s\n" % item)
    
    return participant_comments
```

```python
get_comments(study='EventsAdapt')
```
