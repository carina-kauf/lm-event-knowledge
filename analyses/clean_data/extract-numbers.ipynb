```python
import pandas as pd
```

# EventsAdapt

```python
data_ea = pd.read_csv("clean_EventsAdapt_SentenceSet.csv")
data_ea
```

## number of items per condition

```python
summe = 0
for tt in ["AI", "AAN", "AAR"]:
    sub_dat = data_ea.loc[data_ea["TrialType"] == tt]
    print(f"{tt} | {int(len(sub_dat)/4)}")
    summe += len(sub_dat)
print(summe)
print(summe/4)
```

## number of synonym items

```python
summe = 0
syn_data = data_ea.loc[~data_ea["SynonymPair"].isna()]
for tt in ["AI", "AAN", "AAR"]:
    sub_dat = syn_data.loc[syn_data["TrialType"] == tt]
    print(f"{tt} | {int(len(sub_dat)/4)}")
    summe += len(sub_dat)
print(summe)
print(summe/4)
```
