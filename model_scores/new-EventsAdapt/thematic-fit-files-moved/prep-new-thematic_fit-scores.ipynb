# Grab sentences from old score file & add scores from new file!

```python
import pandas as pd
```

```python
old_prod = "newsentences-EventsAdapt.update-model.TF-prod_OLD.txt"
old_add = "newsentences-EventsAdapt.update-model.TF-add_OLD.txt"
```

```python
new_scores_prod = "newsentences_EventsAdapt.n200.m20.prod.lmi.scores.txt"
new_scores_add = "newsentences_EventsAdapt.n200.m20.add.lmi.scores.txt"
```

# Prod DF

```python
sentence_df_prod = pd.read_csv(old_prod, sep="\t",header=None)
sentence_df_prod.columns = ["index", "sentence", "score"]
sentence_df_prod
```

```python
new_scores_df_prod = pd.read_csv(new_scores_prod, sep="\t")
new_scores_df_prod
```

```python
new_prod_scores = list(new_scores_df_prod["cosine_similarity(filler,prototype)"])
new_prod_scores[:5]
```

```python
new_prod_df = sentence_df_prod.copy()
new_prod_df["score"] = new_prod_scores
new_prod_df
```

```python
new_prod_df.to_csv("newsentences-EventsAdapt.update-model.TF-prod.txt",sep="\t",header=False,index=None)
```

# Addition DF

```python
sentence_df_add = pd.read_csv(old_add, sep="\t",header=None)
sentence_df_add.columns = ["index", "sentence", "score"]
sentence_df_add
```
